## Q1: Simulate aliasing of sinusoidal signals

For this problem you will simulate the phenomena of aliasing.  This will give you pracitice in making synthetic time series, subsetting data, and plotting your results.

**1** To start, make a time vector that goes from 0 to 120 s, with 10000 entries in it.  Then make a synthetic time series that has two close in frequency sine waves:

$$ x = A_1 \cos (\omega t) + A_2 \cos ((\omega_1+\delta \omega ) t) $$

where $\omega$ is the "carrier" frequency, and $\delta \omega << \omega$ is the difference in frequencies between the two waves.  For simplicty, you could make $A_1=A_2=1. [V]$.  Feel free to choose an interesting $\omega$.

<hr>

```python
# You will almost always need to import these:  Make sure to run this cell!
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
arraytime = np.linspace(0.,120.,10000.)
w = 2
dw = 0.001
x1 = 1.*np.cos(arraytime*w)
x2 = 1.*np.cos(arraytime*(w+dw))
X = x1 + x2
```

```python
# YOUR CODE HERE
fig=plt.figure(figsize=(10,15))
ax = fig.add_subplot(3,1,3)
ax.plot(arraytime,x1,label='X1')
ax.plot(arraytime,x2,label='X2')
ax.plot(arraytime,X,label="X1 + X2")
ax.set_xlabel('Time [s]')
ax.set_ylabel('X [V]')
ax.set_title('Aliasing of Sinusoidal Signals')
ax.legend(loc=1)
```

## Q2: Demonstrate the central limit theorem for a very unfair coin flip.

Consider a coin that is twice as likely to come up heads as tails. To put a number on it, call tails $x=-1/2$ and heads $x=+1/2$.  

**1** What are the mean, $\mu_x$, and variance $\sigma_x^2$ of this process?

<hr>

YOUR ANSWER HERE

$\mu_x$ = $(2/3)(1/2) + (1/3)(-1/2) = 1/6 = 0.166667$

$\sigma_x^2$ = $(2/3)[(1/2) - (1/6)]^2 + (1/3)[(-1/2) - (1/6)]^2 = 2/9 = 0.222222$

```python
# YOUR CODE HERE

import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-1.,1.,0.0000001)
P = x*0.
P[x>-0.5]=1./3.
P[x>0.5]=1
fig=plt.figure();ax=fig.add_subplot(1,1,1)
ax.plot(x,P)
ax.set_xlabel('x - Coin Side Value')
ax.set_ylabel('P(x) - Probability of Flip Value Less Than x');ax.set_ylim([-0.1,1.1])
ax.set_title('Cumulative Distribution for Very Unfair Coin Flip')
```

```python
# YOUR CODE HERE
def cointoss(N, doprint=False): #prints the heads/tails values and returns the sample mean
    total =0
    results =[]
    for i in range (1,N+1,1):
        x=np.random.rand()
        if(x<1./3.):
            result = -1./2.
        else:
            result = 1./2.
        total=total+result
        results.append(result)
        mean = total/N
    if(doprint==True):
        print results
        print 'mean:' 
        print total/N  
    return total/N
x=cointoss(5,True) #demonstrates that function works
```

**3** Write a function using the `np.random.rand` function to simulate the coin flip.  It shoudl return $+1/2+$ twice as often as $-1/2$.  Demonstrate that your function works.

```python
# YOUR CODE HERE
twoflips=np.zeros(2000)
for i in range (0,2000,1):
    twoflips[i]=cointoss(2)

sixteenflips=np.zeros(2000)
for i in range (0,2000,1):
    sixteenflips[i]=cointoss(16)

sixtyfourflips=np.zeros(2000)
for i in range (0,2000,1):
    sixtyfourflips[i]=cointoss(64)
    
twofiftysixflips=np.zeros(2000)
for i in range (0,2000,1):
    twofiftysixflips[i]=cointoss(256)
```

**4** Perform $N_{flip}$ and take the sample mean $\overline{x}$ of the values.  Do this 2000 times, and show that as $N_{flip}\to\infty$ the distribution of $\overline{x}$ approaches a Normal distribution by plotting the histograms of the means for each.  For $N_{flip}$ choose 2, 16, 64, 256. 

Comment on how the convergence of $\overline{x}$ for the unfair coin compares to the convergence of the sample mean of the dice roll from the notes.

```python
# YOUR CODE HERE
fig=plt.figure();ax=fig.add_subplot(1,1,1); 
ax.hist(twoflips,30)
ax.set_title('Number of Flips = 2')
ax.set_xlabel('$\overline{x}\ \  N=2$')
ax.set_ylabel('Occurences')

fig=plt.figure();ax=fig.add_subplot(1,1,1); 
ax.hist(sixteenflips,30)
ax.set_title('Number of Flips = 16')
ax.set_xlabel('$\overline{x}\ \  N=16$')
ax.set_ylabel('Occurences')


fig=plt.figure();ax=fig.add_subplot(1,1,1); 
ax.hist(sixtyfourflips,30)
ax.set_title('Number of Flips = 64')
ax.set_xlabel('$\overline{x}\ \  N=64$')
ax.set_ylabel('Occurences')


fig=plt.figure();ax=fig.add_subplot(1,1,1); 
ax.hist(twofiftysixflips,30)
ax.set_title('Number of Flips = 256')
ax.set_xlabel('$\overline{x}\ \  N=256$')
ax.set_ylabel('Occurences')
```

YOUR ANSWER HERE

The $\overline{x}$ distribution appears to converge faster for the coin flip than for the dice roll from the notes, in that the width of the distribution for the coin toss is thinner than the width of the distribution for the dice roll. Both distributions appear to converge to narrower and narrower peaks centered at the sample mean as N is increased. The distribution for the coin flip may converge faster than the distribution for the dice roll because there are six possible outcomes of rolling the dice and only two possible outcomes from flipping the coin.
